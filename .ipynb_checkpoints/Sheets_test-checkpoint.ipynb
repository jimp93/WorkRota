```python
import gspread
from pprint import pprint
import datetime as dt
from googleapiclient.discovery import build
```

```python
from oauth2client.service_account import ServiceAccountCredentials
today = dt.date.today()
today
```

```python
scope=["https://spreadsheets.google.com/feeds",'https://www.googleapis.com/auth/spreadsheets',"https://www.googleapis.com/auth/drive.file","https://www.googleapis.com/auth/drive"]
```

```python
cal_scope=["https://www.googleapis.com/auth/calendar"]
```

```python
creds = ServiceAccountCredentials.from_json_keyfile_name("creds.json", scope)
```

```python
cal_creds = ServiceAccountCredentials.from_json_keyfile_name("creds.json", cal_scope)
```

```python
service = build('calendar', 'v3', credentials=cal_creds)
```

```python
client = gspread.authorize(creds)
```

```python
future_list=[]
work_sdates=[]
work_fdates=[]

def make_rota_list(sheet, mon_date):
    times_rows=[]
    full_wsheet = sheet.get_all_values()
    times_rows = full_wsheet[2:6]
    for tr in times_rows:
        rota_hr = tr[0]
        sdatetime = dt.datetime.strptime(rota_hr, '%I:%M %p')
        fdatetime = sdatetime + dt.timedelta(hours=7, minutes=30)
        stime= sdatetime.time()
        ftime = fdatetime.time()
        tr.pop(6)
        tr=tr[1:]
        for t in range(7):
            if 'jwp' in tr[t].lower():
                rot_date = mon_date + dt.timedelta(days=t)
                ev_stime = dt.datetime.combine(rot_date, stime)
                ev_ftime = dt.datetime.combine(rot_date, ftime)
                ev_st_iso = ev_stime.isoformat()
                ev_ft_iso = ev_ftime.isoformat()
                work_sdates.append(ev_st_iso)
                work_fdates.append(ev_ft_iso)
        global dates_list
        dates_zip = zip(work_sdates, work_fdates)
        dates_list = list(dates_zip)

for ra in range(1,30):
    sheet = client.open("London newsdesk rotas").get_worksheet(ra)
    date_row = sheet.row_values(1)
    str_o_split = date_row[0].split("-")
    str_m_split = str_o_split[0].split(" ")
    month = str_m_split[0][:3]
    str_date = month + " " + str_m_split[1] + " " + str_o_split[1][-4:]
    d = dt.datetime.strptime(str_date, '%b %d %Y')
    mon_date = d.date()
    difference = mon_date - today
    if difference > dt.timedelta(days=-7):
        make_rota_list(sheet, mon_date)
    else:
        break
```

```python
for tim in dates_list:
    event = {
        'summary': 'WORK',
        'start': {
        'dateTime': tim[0] + '+01:00',},
        'end': {
        'dateTime': tim[1] + '+01:00',},
    }
    event = service.events().insert(calendarId='cgejrm0nfeibsstmlr74s2huq8@group.calendar.google.com', body=event).execute()
```
